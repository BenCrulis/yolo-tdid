```python
import os
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import re
```

```python
result_orig = Path('../yolo_tdid/results/')
result_wc = Path('results_wc_gqa_partial')
# result_wc = Path('results_wc_eps')
```

```python
list(result_wc.iterdir())[0]
```

```python
def load(path):
    files = path.iterdir()
    name_re = re.compile(r"yolow-(?P<size>.)_sighted-(?P<sighted>.+)_clutter-(?P<clutter>.+)_hand-(?P<hand>.+)_n-(?P<n>\d+)_k-(?P<k>\d+)_bias-.+_augment-(?P<augment>True|False+)(_use_wc-(?P<wc>.+))?_seed-(?P<seed>\d+).*")

    dfs = []

    for f in files:
        m = name_re.match(f.stem)
        if m is None:
            print(f"Skipping {f}")
            continue
        try:
            df = pd.read_csv(f)
            df['size'] = m.group("size")
            df['train blind'] = m.group("sighted") == "False"
            df['train clutter'] = m.group("clutter") == "True"
            df['train hand'] = m.group("hand") == "True"
            df['n_objects'] = int(m.group("n"))
            df['k_examples'] = int(m.group("k"))
            df['augment'] = m.group("augment") == "True"
            df['WC'] = m.group("wc") == "True"
            dfs.append(df)
        except Exception as e:
            print(e)
            print("skipping file")

    df = pd.concat(dfs, ignore_index=True)
    return df
```

```python
df_base = load(result_orig)
```

```python
df_wc = load(result_wc)
```

```python
df_a = df_base[df_base["augment"] & (df_base["train hand"] == False)]
```

```python
acc_base = df_a.groupby(['size', 'n_objects', 'k_examples', 'augment', 'seed'])["correct"].mean()\
    .groupby(['size', 'n_objects', 'k_examples', 'augment']).mean()
```

```python
acc_wc = df_wc.groupby(['size', 'n_objects', 'k_examples', 'augment', 'seed'])["correct"].mean()\
    .groupby(['size', 'n_objects', 'k_examples', 'augment']).mean()
```

```python
diff = acc_wc - acc_base
```

```python
diff
```

```python
pivoted = diff.reset_index().pivot_table(values=["correct"], index=['n_objects', 'k_examples'], columns=["augment", "size"])
pivoted = pivoted.reorder_levels([1, 2, 0], axis=1)
pivoted = pivoted.sort_index(axis=1, level=[2], sort_remaining=False, ascending=False)\
    .sort_index(axis=1, level=[1], ascending=False, sort_remaining=False)\
    .sort_index(axis=1, level=[0], sort_remaining=False)
pivoted
```
